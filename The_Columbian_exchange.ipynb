# Smallpox in America, one byproduct of the Columbian exchange

<p style='text-align: justify;'>  
    The expression “Colombian exchange” refers to the mutual effects that followed the contact of Spaniards mainly, and Europeans and Africans by extension, with the indigenous populations of America. It corresponds to an exchange of animals, plants, technologies, but also of many diseases.
Our choice of study focused on smallpox disease, which was first introduced into South America by the conquistadors and was then one of the major causes of the pre-Columbian populations's decimation.
</p>

#### Convert latitudes and longitudes into graphic coordinates

```python
import math
from ast import literal_eval
def merc(Coords):
    Coordinates = literal_eval(Coords)
    lat = Coordinates[0]
    lon = Coordinates[1]
    
    r_major = 6378137.000
    x = r_major * math.radians(lon)
    scale = x/lon
    y = 180.0/math.pi * math.log(math.tan(math.pi/4.0 + 
        lat * (math.pi/180.0)/2.0)) * scale
    return (x, y)

def make_tuple_str(x, y):
    t = (x, y)
    return str(t)


df['coords'] = df.apply(lambda x: make_tuple_str(x['Latitude'], x['Longitude']), axis = 1)
df['coords_latitude'] = df['coords'].apply(lambda x: merc(x)[0])
df['coords_longitude'] = df['coords'].apply(lambda x: merc(x)[1])

df2['start_coords_lat'] = [df2['Coordinates of departure (latitude, longitude)'][i][0] for i in range(len(data2['Lead sailor']))]
df2['end_coords_lat'] = [df2['Coordinates of arrival (latitude, longitude)'][i][0] for i in range(len(data2['Lead sailor']))]

df2['start_coords_long'] =[df2['Coordinates of departure (latitude, longitude)'][i][1] for i in range(len(data2['Lead sailor']))]
df2['end_coords_long'] =[df2['Coordinates of arrival (latitude, longitude)'][i][1] for i in range(len(data2['Lead sailor']))]


df2['start_coords'] = df2.apply(lambda x: make_tuple_str(x['start_coords_lat'], x['start_coords_long']), axis = 1)
df2['end_coords'] = df2.apply(lambda x: make_tuple_str(x['end_coords_lat'], x['end_coords_long']), axis = 1)


df2['start_coords_lat'] = df2['start_coords'].apply(lambda x: merc(x)[0])
df2['end_coords_lat'] = df2['end_coords'].apply(lambda x: merc(x)[0])


df2['start_coords_long'] = df2['start_coords'].apply(lambda x: merc(x)[1])
df2['end_coords_long'] = df2['end_coords'].apply(lambda x: merc(x)[1])

```

## The Columbian exchange, an example of spatial approach to diseases (6000 signs text)

<p style='text-align:justify;'> In 1492, Christopher Columbus did not "discover" America strictly speaking: this land had long been inhabited by natives, and the Vikings (among others) had already « bridged » the gap between the American continent and Eurasia. However, Columbus’s expedition did paved the way from what would become the strong connection and the interdependency between the so-called Old and New Worlds. From this day on, transfers of all imaginable things, living beings, human practices, and ideas have increasingly intensify and produced irreversible changes on both continents. These exchanges, which particularly affected the American fauna and flora, have been deeply investigated by the environmental historian Alfred Crosby who gave them the name of Colombian exchange in a 1972’s publication entitled The Columbian exchange: Biological and Cultural Consequences of 1492. Indeed, America was full of living beings new to Europeans who were therefore all at once frightened, greatly amazed and highly intrigued by their encounters. The sailors brought back some of their discoveries to the Old World to be studied or exhibited. Likewise, as settlers needed Old World cattle, crops, and other animals to thrive and feel in control of the land, they started to import those to the New World. Very soon, flux intensified and trade was permanently established. America’s isolation broke. New supplies became - more or less - accessible to all continents through the Columbian exchange, although trades were often made at natives’ expense. </p>
<p style='text-align:justify;'>However, this is only the tip of the iceberg. As demonstrated by environmental historians and spatial epidemiologists, America and Europe did not only exchange animals and crops but also germs and diseases, and the difference between populations was not only cultural: it was also a contrast between immunities. As there had been almost no contact between those two parts of the  world during thousands of years, each had developed its own diseases, had biologically adapted to resist those specific diseases, had learned to control the spread and had created social responses to the symptoms of those specific diseases they had learned to know. As a result, the impact of disease exchanges was huge. They were even more dramatic among natives whose unique and singular  homogeneous biology made excessively vulnerable to European germs (« no population of comparable size has remained so uniform after expanding, in whatever time has been involved, over such a large area. » T. D. Stewart, 1960). Thus, Europeans brought syphilis back from America (although this thesis is controversial) when they themselves spread smallpox, measles and typhoid in America. </p>

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime

events = [" Earliest evidence of smallpox : \n mummified remains of Egyptian Pharaoh Ramses V \n show telltale pockmarks on his skin",
          "Smallpox is introduced into \nthe Caribbean island of Hispaniola ",
          "Edward Jenner successfully tested in humans \na cowpox vaccine against smallpox", 
          "The World Health Organization launched \na worldwide eradication campaign \nbased on vaccination ",
          "Last known case of smallpox on the American continent",
          "Last known case of smallpox worldwide (Somalia)"]

dates = [-1157, 1509, 1796, 1967, 1971,1977]

fig, ax = plt.subplots(figsize=(10, 7), constrained_layout=True)

for event, i , date in zip(events[::-1], range(len(dates)), dates[::-1]) :
    ax.annotate("{} ".format(date)+"\n"+event, xy=(0, i), 
                fontsize=10,
                horizontalalignment="left" if i%2==0 else "right")

ax.set_title("3000 years of smallpox from its first detection to its eradication", c="#3776ab", weight=1000)
ax.plot(np.zeros_like(dates), range(len(dates)), "-o",
        color="k", markerfacecolor="w") 

ax.axis("off")
ax.margins(y=0.2)
text = "Caption : Smallpox’s history cannot be reduced to Europeans’ contamination of the American continent. First detected in Egypt, it was spread back \nand forth in Eurasia and Africa well before the Columbian exchange. To this date, smallpox is the only human disease successfully eradicated by man."
fig.text(0,0,text, fontsize = 12,  multialignment= "left")
plt.show()
```

<p style='text-align:justify;'>Of all the diseases involuntarily brought by Europeans to America, smallpox may have been the deadliest (Crosby, 1972). Both Variola major and Variola minor arrived in America. Passed from human to human, smallpox was very contagious as it could transmit through the air by oral and nasal droplets or through contact with erupted pustules. After 12 days of incubation, infected patients would suffer from fever, fatigue, vomiting and intense pain preceding a more severe and dangerous eruptive stage. At this point of the disease, victims of smallpox were easily identified as their skin was quickly covered with pustules or fluid-filled blisters, scarring them. Throughout the whole month-long process, sick people were contagious. Finally, those who did not die of smallpox would become at least partially immune to it. </p>
<p style='text-align:justify;'>The timeline shows that smallpox was introduced to America a few years only after Christopher Columbus’s first landing on the Caribbean island of Hispaniola in 1492. Indeed, some inhabitants of this island were infected as soon as 1509. The interactive map sheds light on the correlation between Europeans presence on American soil and the fast spreading of smallpox all over the continent. As mentioned, smallpox was a virulent disease and infected people were contagious even before the first symptoms appeared, while they were still able to travel. Contrary to stereotypical narratives, natives’ tribes were in a dense network of exchange and communication with one another long before the arrival of the Europeans. Therefore, smallpox traveled quickly throughout the whole continent.</p>
<p style='text-align:justify;'> However, that being said, one should keep in mind that the link between disease spreading and intercontinental migration dates back well before the Columbian exchange. Indeed, the timeline shows that the first known case of smallpox was detected in Egypt in 1157 BC. Centuries before  the Columbian Exchange, European trade with another continent (in this case, Africa) led to the spreading of smallpox in Europe. Trade has long been a vector of diseases. </p>

--------

<p style='text-align:justify;'>The study of smallpox in post-columbian America is of great interest because it is conducive to Alfred Crosby’s notion of "virgin soil epidemic » i.e. epidemics affecting populations that “had no previous contact with the diseases that strike them and are therefore immunologically almost defenceless” (Crosby, 1972). As natives had not yet been in contact with smallpox, they were extremely vulnerable to it whereas Spaniards showed far more resistance thanks to the evolution of European immune systems and acquired immunity. Natives’ immuno-naïvity therefore explains the numerous outbreaks following Europeans’ settlings that we can see on the interactive map.</p>
<p style='text-align:justify;'>Moreover, the study of fatality case rates (CFR) allows to better understand the evolution of smallpox and the different immunity of populations affected by the disease. The CFR is the proportion of deaths compared to the number of diseased. It should logically decrease when populations’ immunity is increasing. The second map shows this decrease. The intensification of trade throughout the world led to progressive biological homogeneity, everyone developing immunity against the diseases they share. This is why the fatality case rate of smallpox decreases over time. Added to the natural development of immunity, the development of a vaccine against smallpox in 1796 (according to the timeline) explains the decrease of fatality case rates we can see on the map. This vaccine was the subject of a worldwide campaign against smallpox organized by WHO in 1967. Thanks to this campaign, smallpox was the first and only disease eradicated by humans. The last case known on American continent and in the world, were respectively discovered in 1971 and 1977.</p>
 
-----------

<p style='text-align:justify;'>Smallpox helped by other diseases decimated the local population. 80% to 95% of Amerindians were decimated by 1650 (Newson, 2001). This demographic catastrophe had political consequences. Diseases such as smallpox were a very efficient weapon against natives. Smallpox epidemics decimated by turns the Aztec and the Incas population and army. This facilitated European expeditions and colonization. Those easy conquests reinforced Europeans’ conviction of a divine mission of civilization through colonization. Therefore, diseases from the Columbian Exchange had an important role in the colonization of the American continent and its irreversible demographic and ecological consequences.</p>
