#### Convert latitudes and longitudes into graphic coordinates

```python
import math
from ast import literal_eval
def merc(Coords):
    Coordinates = literal_eval(Coords)
    lat = Coordinates[0]
    lon = Coordinates[1]
    
    r_major = 6378137.000
    x = r_major * math.radians(lon)
    scale = x/lon
    y = 180.0/math.pi * math.log(math.tan(math.pi/4.0 + 
        lat * (math.pi/180.0)/2.0)) * scale
    return (x, y)

def make_tuple_str(x, y):
    t = (x, y)
    return str(t)

df['coords'] = df.apply(lambda x: make_tuple_str(x['Latitude'], x['Longitude']), axis = 1)
df['coords_latitude'] = df['coords'].apply(lambda x: merc(x)[0])
df['coords_longitude'] = df['coords'].apply(lambda x: merc(x)[1])
df
```

### Display a map showing smallpox epidemics and transatlantic crossings over time 

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.tile_providers import CARTODBPOSITRON_RETINA, get_provider, Vendors 
from bokeh.models import HoverTool, Label, ColumnDataSource
from datetime import date
from bokeh.models import CustomJS, DateSlider, Slider, Arrow, VeeHead
from bokeh.layouts import column


hover1 = HoverTool(tooltips = [("Location (actual name)", "@desc"),
                             ("Smallpox epidemics","@start_year - @end_year " )], names = ['cir'])

hover2 = HoverTool(tooltips = [("Year of arrival on the American continent", "@crossing_year"),
                             ("Lead sailor","@sail " )], names = ['seg'])

hover3 = HoverTool(tooltips = [('','@desc3')], names = ['dia'])

p = figure(plot_width=800, plot_height=550,
           x_range=(min(df['coords_latitude'])-1.5e+05, max(df['coords_latitude']+1.5e+06)), 
           y_range=(min(df['coords_longitude'])-1e+06,max(df['coords_longitude'])+1e+06),
           title="Trouver un titre",
           x_axis_type="mercator", 
           y_axis_type="mercator", tools = "pan, wheel_zoom, box_zoom, reset")

tile_provider = get_provider(Vendors.CARTODBPOSITRON_RETINA)
p.add_tile(tile_provider)
p.add_tools(hover1, hover2, hover3)


source_epidemics = ColumnDataSource(data=dict(
    x = df['coords_latitude'],
    y = df['coords_longitude'],
    dim = [0 for i in range(len(df['coords_longitude']))],
    start_year = df['Start Year'],
    end_year = df['End Year'],
    desc = df['Location (actual name)']
,))


source_travel = ColumnDataSource(data=dict(
    x_start = df2['start_coords_lat'],
    y_start = df2['start_coords_long'],
    x_end = df2['end_coords_lat'],
    y_end = df2['end_coords_long'],
    vis = [0 for i in range(len(df2['end_coords_lat']))],
    crossing_year = df2['Year of arrival on the American continent'],
    sail = df2['Lead sailor']
,))


source_last_case = ColumnDataSource(data=dict(
    x = [merc('(-10.3333333,-53.2)')[0]], 
    y = [merc('(-10.3333333,-53.2)')[1]],
    last_case_year = [1971],
    desc3 = ["April 19, 1971"],
    dim3 = [0.]
,))


callback = CustomJS(args=dict(source1=source_epidemics, source2=source_travel, source3=source_last_case), code="""
    var value = cb_obj.value;
    
    var data1 = source1.data;  
    var start_year = data1['start_year']; 
    var end_year = data1['end_year'];
    var dims = data1['dim'];
    for (var i = 0; i < dims.length; i++) {
        if ( start_year[i] <= value ) {
            if (end_year[i]>= value) {dims[i]=30 }
            else {dims[i]=13 }
            }
        else {dims[i]=0 }
    }
        
    var data2 = source2.data;
    var visi = data2['vis'];
    var cr_year = data2['crossing_year']
    for (var i = 0; i < visi.length; i++) {
        if (cr_year[i]<= value) { visi[i] = 1}
        else {visi[i]=0 }
    }
    
    var data3 = source3.data;
    var dims3 = data3['dim3'];
    var last_case_year = data3['last_case_year'];
    for (var i = 0; i < dims3.length; i++) {
        if (last_case_year[i] <= value ) {dims3[i] = 20}
        else {dims3[i]=0 }
    }
    
    source3.change.emit(); 
    source2.change.emit(); 
    source1.change.emit();
""")

p.circle('x', 'y', legend_label="Major smallpox epidemics in America", 
         color= "red", fill_color="red", alpha=0.6, hover_color='orange',
         size ='dim',
         name = 'cir',
         source = source_epidemics)



p.segment(x0="x_start", y0="y_start", x1="x_end", y1="y_end", 
          line_color="grey", line_width=.5, line_dash='dashed', hover_color='orange',
          line_alpha = 'vis', 
          name = 'seg',
          source = source_travel)


p.diamond('x', 'y', legend_label="Last known case of smallpox on the continent",
         color= "green", hover_color='grey',
         size = 'dim3',
         name = 'dia',
         source = source_last_case 
         )

#legend_label="Brazil, April 19, 1971" +"\nLast known case of smallpox (variola minor) on the continent"






slider = Slider(start=1492, end=1975, value=1492, step=1, title="Date")

slider.js_on_change('value', callback)


layout = column(slider, p)
output_notebook()

show(layout)
```
